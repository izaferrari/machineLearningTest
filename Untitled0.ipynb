```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
from keras.utils import to_categorical
from tensorflow.keras.models import load_model
```

```python
dataset = keras.datasets.cifar10

((imagens_treino, identificacoes_treino), (imagens_teste, identificacoes_teste)) = dataset.load_data()
```

```python
total_classificacoes = 10
nomes_classificados = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
for imagem in range(10):
  plt.subplot(2, 5, imagem+1)
  plt.imshow(imagens_treino[imagem])
  plt.title(identificacoes_treino[imagem])
```

```python
imagens_treino = imagens_treino/float(250)
```

```python
modelo = keras.Sequential([
    keras.layers.Flatten(input_shape=(32, 32, 3)),
    keras.layers.Dense(256, activation=tf.nn.relu),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
etiquetas_treino_encoded = to_categorical(identificacoes_treino, total_classificacoes)
```

```python
modelo.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
testando = modelo.fit(imagens_treino, etiquetas_treino_encoded, epochs=5, validation_split=0.2)
```

```python
plt.plot(testando.history['accuracy'])
```

```python
plt.plot(testando.history['accuracy'])
plt.plot(testando.history['val_accuracy'])
plt.xlabel('epocas')
plt.ylabel('acuracia')
plt.legend(['treino', 'validaçao' ])
```

```python
plt.plot(testando.history['loss'])
plt.plot(testando.history['val_loss'])
plt.xlabel('perdas por epocas')
plt.ylabel('perda')
plt.legend(['treino', 'validaçao' ])
```

```python
modelo.save('primeiroModelo')
modelo_salvo = load_model('primeiroModelo')
```

```python
testes = modelo.predict(imagens_teste)
print('resultado teste:', np.argmax(testes[3]))
print('numero imagem do teste: ', identificacoes_teste[3])
```

```python
testes_modelo_salvo = modelo_salvo.predict(imagens_teste)
print('resultado do teste do modelo salvo: ', np.argmax(testes_modelo_salvo[3]))
print('numero da imagem do teste: ', identificacoes_teste[3])
```
